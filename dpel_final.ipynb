```python
import pandas as pd

# Load the dataset
file_path = '/content/DPDEL-FORM (Responses) - Form responses 1.csv'
df = pd.read_csv(file_path)

# Display the first five rows
print("First 5 rows of the dataset:")
display(df.head())

# Check data info
print("\nDataset Info:")
display(df.info())

# View summary statistics
print("\nSummary Statistics:")
display(df.describe(include='all'))

# Mention dataset shape
print("\nDataset shape:", df.shape)
```

###  Step 2: Data Cleaning & Imputation (CO2)

```python
# Rename columns
new_column_names = {
    'Timestamp': 'timestamp',
    'DATE': 'date',
    'Age': 'age',
    'Heart Rate(BPM)': 'heart_rate_bpm',
    'Blood Oxygen Level (%)--(only numbers)': 'blood_oxygen_level_percentage',
    'Sleep Duration (Hours)': 'sleep_duration_hours',
    ' Sleep Quality': 'sleep_quality',
    'Body Weight (in KGs)': 'body_weight_kgs',
    'Activity Level': 'activity_level',
    'Screen Time (Hourly)': 'screen_time_hours_daily',
    'Meal Regularity': 'meal_regularity',
    'Sleep Consistency (same bedtime daily?)': 'sleep_consistency',
    'Step Count(Daily)': 'step_count_daily',
    'Stress Level (Self-Report)  (1 = Relaxed, 10 = Extremely Stressed)': 'stress_level',
    ' Gender': 'gender',
    ' Profession/Role': 'profession_role'
}
df.rename(columns=new_column_names, inplace=True)

# Clean up remaining special characters and spaces
df.columns = df.columns.str.strip().str.replace('[^a-zA-Z0-9_]', '', regex=True).str.lower()

# Explicitly rename the stress level column after general cleaning
df.rename(columns={'stresslevelselfreport1relaxed10extremelystressed': 'stress_level'}, inplace=True)


# Display the renamed columns
print("Columns after renaming:")
print(df.columns)
```

```python
# Handle missing values

# Impute numeric columns with the median
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
for col in numeric_cols:
    if df[col].isnull().any():
        median_value = df[col].median()
        df[col] = df[col].fillna(median_value) # Avoid inplace=True with chained assignment
        print(f"Imputed missing values in '{col}' with median ({median_value})")

# Impute categorical columns with the mode
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    if df[col].isnull().any():
        mode_value = df[col].mode()[0]
        df[col] = df[col].fillna(mode_value) # Avoid inplace=True with chained assignment
        print(f"Imputed missing values in '{col}' with mode ({mode_value})")


# Verify that there are no more missing values
print("\nMissing values after imputation:")
print(df.isnull().sum())
```

###  Step 3: Exploratory Data Analysis (EDA)

```python
# Drop duplicates
initial_rows = df.shape[0]
df.drop_duplicates(inplace=True)
rows_after_dropping_duplicates = df.shape[0]
print(f"\nNumber of rows before dropping duplicates: {initial_rows}")
print(f"Number of rows after dropping duplicates: {rows_after_dropping_duplicates}")
```

```python
# Handle outliers using the IQR method

# Identify numeric columns for outlier detection
numeric_cols_for_outliers = df.select_dtypes(include=['float64', 'int64']).columns.tolist()

# Exclude 'age' and 'stress_level' from outlier removal as they represent reported values
if 'age' in numeric_cols_for_outliers:
    numeric_cols_for_outliers.remove('age')
if 'stress_level_self_report_1_relaxed_10_extremely_stressed' in numeric_cols_for_outliers:
    numeric_cols_for_outliers.remove('stress_level_self_report_1_relaxed_10_extremely_stressed')
# Exclude sleep_quality_numeric and sleep_efficiency as they are newly created and might have NaNs/Infs before proper handling
if 'sleep_quality_numeric' in numeric_cols_for_outliers:
    numeric_cols_for_outliers.remove('sleep_quality_numeric')
if 'sleep_efficiency' in numeric_cols_for_outliers:
    numeric_cols_for_outliers.remove('sleep_efficiency')
if 'activity_to_stress_ratio' in numeric_cols_for_outliers:
    numeric_cols_for_outliers.remove('activity_to_stress_ratio')

print(f"\nHandling outliers for the following columns: {numeric_cols_for_outliers}")

for col in numeric_cols_for_outliers:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Remove outliers
    outlier_count = df[(df[col] < lower_bound) | (df[col] > upper_bound)].shape[0]
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

    print(f"Removed {outlier_count} outliers from '{col}'")

print("\nDataset shape after outlier removal:", df.shape)
```

## Feature engineering

### Subtask:
Create new derived features such as `Sleep_Efficiency` and `Activity_to_Stress_Ratio` based on the existing columns.


```python
# Create Sleep_Efficiency feature
# Note: Sleep Quality is categorical, so we cannot directly divide.
# We need to map sleep quality to numerical values first.
# Let's assume 'Good' is 3, 'Average' is 2, and 'Poor' is 1 for this calculation.
sleep_quality_mapping = {'Good': 3, 'Average': 2, 'Poor': 1} # Use correct capitalization based on original data


print("Unique values in 'sleepquality' before mapping:", df['sleepquality'].unique()) # Diagnostic print
print("Value counts in 'sleepquality' before mapping:\n", df['sleepquality'].value_counts(dropna=False)) # Diagnostic print


df['sleep_quality_numeric'] = df['sleepquality'].map(sleep_quality_mapping)

print("\nUnique values in 'sleep_quality_numeric' after mapping:", df['sleep_quality_numeric'].unique()) # Diagnostic print
print("Value counts in 'sleep_quality_numeric' after mapping:\n", df['sleep_quality_numeric'].value_counts(dropna=False)) # Diagnostic print


# Handle potential NaN values after mapping (if any sleep quality was not in the mapping or was originally missing)
# Impute with the mode instead of the median
mode_series = df['sleep_quality_numeric'].mode()
if not mode_series.empty:
    mode_sleep_quality_numeric = mode_series[0]
    df['sleep_quality_numeric'] = df['sleep_quality_numeric'].fillna(mode_sleep_quality_numeric)
else:
    print("\nWarning: 'sleep_quality_numeric' mode is empty. Imputing with a default value (e.g., 0) to avoid errors.")
    df['sleep_quality_numeric'] = df['sleep_quality_numeric'].fillna(0)


# Create Sleep_Efficiency, handle division by zero if necessary (though with the mapping, this is unlikely)
# Add a small epsilon to the denominator to prevent division by zero
df['sleep_efficiency'] = df['sleep_duration_hours'] / (df['sleep_quality_numeric'] + 1e-6)


# Create Activity_to_Stress_Ratio feature, handle division by zero if necessary
# Check if stress_level has any zero values before dividing
print("\nColumns in DataFrame before accessing 'stress_level_self_report_1_relaxed_10_extremely_stressed':", df.columns) # Diagnostic print
if (df['stress_level'].min() == 0):
    print("Warning: 'stress_level' contains zero values. Adding a small epsilon to avoid division by zero.")
    df['activity_to_stress_ratio'] = df['step_count_daily'] / (df['stress_level'] + 1e-6)
else:
    df['activity_to_stress_ratio'] = df['step_count_daily'] / df['stress_level']

# Display the head of the dataframe with the new features
display(df[['sleep_duration_hours', 'sleepquality', 'sleep_quality_numeric', 'sleep_efficiency', 'step_count_daily', 'stress_level', 'activity_to_stress_ratio']].head())
```

## Model implementation

### Subtask:
Train a Logistic Regression, Decision Tree Classifier, and Random Forest Classifier on the training data.


**Reasoning**:
Train Logistic Regression, Decision Tree, and Random Forest models on the training data as requested.



```python
# ===============================================
#  1. Data Preparation for Autoencoder
# ===============================================

# Identify numerical and categorical columns
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
categorical_cols_to_encode = [
    'sleepquality', 'activity_level', 'meal_regularity', 'sleepconsistencysamebedtimedaily',
    'gender', 'profession_role', 'sleep_quality_numeric'
]

# One-hot encode categorical columns
df_encoded_categorical = pd.get_dummies(df[categorical_cols_to_encode], drop_first=True)
df_numerical = df[numerical_cols].drop(columns=['timestamp', 'date', 'stress_level'], errors='ignore') # Exclude timestamp, date, and target

# Concatenate encoded and numerical features
autoencoder_input_data = pd.concat([df_numerical, df_encoded_categorical], axis=1)

# Ensure column names are strings and reset index for Keras compatibility
autoencoder_input_data.columns = autoencoder_input_data.columns.astype(str)
autoencoder_input_data.reset_index(drop=True, inplace=True)


print("Autoencoder input data shape:", autoencoder_input_data.shape)
print("Autoencoder input data dtypes:", autoencoder_input_data.dtypes.unique())


# ===============================================
#  2. Build & Train Autoencoder
# ===============================================

from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
import tensorflow as tf # Ensure tensorflow is imported

input_dim = autoencoder_input_data.shape[1]
latent_dim = input_dim // 2

# Encoder
input_layer = Input(shape=(input_dim,))
encoder_layer_1 = Dense(16, activation='relu')(input_layer)
encoder_layer_2 = Dense(8, activation='relu')(encoder_layer_1)
latent_layer = Dense(latent_dim, activation='relu', name='latent_layer')(encoder_layer_2)

# Decoder
decoder_layer_1 = Dense(8, activation='relu')(latent_layer)
decoder_layer_2 = Dense(16, activation='relu')(decoder_layer_1)
output_layer = Dense(input_dim, activation='linear')(decoder_layer_2)

# Autoencoder Model
autoencoder = Model(inputs=input_layer, outputs=output_layer)
autoencoder.compile(optimizer='adam', loss='mse')

# Diagnostic print before training
print("Starting Autoencoder training...")
print("Input data shape for training:", autoencoder_input_data.values.shape)
print("Input data dtype for training:", autoencoder_input_data.values.dtype)


# Train
autoencoder.fit(
    autoencoder_input_data.values, autoencoder_input_data.values,
    epochs=50, batch_size=32, shuffle=True, validation_split=0.2, verbose=0
)
print("Autoencoder training complete.")


# ===============================================
#  3. Extract Latent Features
# ===============================================

encoder = Model(inputs=autoencoder.input, outputs=autoencoder.get_layer('latent_layer').output)
latent_features = encoder.predict(autoencoder_input_data.values)
latent_features_df = pd.DataFrame(latent_features, index=df.index) # Use original df index
print("Latent features shape:", latent_features_df.shape)


# ===============================================
#  4. Prepare Data for Temporal Model
# ===============================================

# Add 'timestamp' and 'stress_level' back for sorting and target variable
# Keep the original index before sorting
latent_features_df['original_index'] = latent_features_df.index
latent_features_df['timestamp'] = df['timestamp']
latent_features_df['stress_level'] = df['stress_level']
latent_features_df.sort_values(by='timestamp', inplace=True)
latent_features_df.reset_index(drop=True, inplace=True)


def create_sequences(data, seq_len):
    X, y, original_indices = [], [], []
    # Exclude the last 'sequence_length' steps as we cannot create a full sequence starting from them
    for i in range(len(data) - seq_len):
        # Features for the sequence (all columns except 'timestamp', 'stress_level', and 'original_index')
        feature_sequence = data.iloc[i:(i + seq_len)].drop(['timestamp', 'stress_level', 'original_index'], axis=1)
        # Target value (stress level at the end of the sequence)
        target = data.iloc[i + seq_len]['stress_level']
        # Original index corresponding to the target value
        original_idx = data.iloc[i + seq_len]['original_index']


        X.append(feature_sequence.values)
        y.append(target)
        original_indices.append(original_idx)

    return np.array(X), np.array(y), np.array(original_indices)

sequence_length = 7
X_temporal, y_temporal, original_indices_temporal = create_sequences(latent_features_df, sequence_length)
print("Temporal input shape:", X_temporal.shape)


# ===============================================
#  5. Build Temporal GRU + Attention Model
# ===============================================

from tensorflow.keras.layers import GRU, Attention, GlobalAveragePooling1D
import numpy as np # Import numpy
import tensorflow as tf


temporal_input = Input(shape=(X_temporal.shape[1], X_temporal.shape[2]))
gru_out = GRU(64, return_sequences=True)(temporal_input)
gru_out = GRU(32, return_sequences=True)(gru_out)
attention_out = Attention()([gru_out, gru_out])
pooled_out = GlobalAveragePooling1D()(attention_out)
dense1 = Dense(32, activation='relu')(pooled_out)
dense2 = Dense(16, activation='relu')(dense1)
temporal_output = Dense(1, activation='linear')(dense2)

temporal_model = Model(inputs=temporal_input, outputs=temporal_output)
temporal_model.compile(optimizer='adam', loss='mse')

# Convert y_temporal to a TensorFlow tensor with the correct dtype
y_temporal_tf = tf.convert_to_tensor(y_temporal, dtype=tf.float32)


temporal_model.fit(X_temporal, y_temporal_tf, epochs=50, batch_size=32, validation_split=0.2, verbose=1)
print("Temporal model trained.")


# ===============================================
#  6. Extract Temporal Features (Context Vectors)
# ===============================================

feature_extractor = Model(inputs=temporal_model.input, outputs=temporal_model.get_layer(index=-4).output) # Index -4 corresponds to 'pooled_out' layer
temporal_features = feature_extractor.predict(X_temporal)
temporal_features_df = pd.DataFrame(temporal_features, index=original_indices_temporal) # Use original indices
print("Temporal feature shape:", temporal_features_df.shape)


# ===============================================
#  7. Prepare Meta-Predictor Input
# ===============================================

relevant_features = [
    'age', 'gender', 'profession_role', 'body_weight_kgs', 'activity_level', 'meal_regularity',
    'sleepconsistencysamebedtimedaily', 'blood_oxygen_level_percentage', 'screen_time_hours_daily',
    'heartratebpm', 'step_count_daily', 'sleep_quality_numeric', 'sleep_efficiency', 'activity_to_stress_ratio'
]

# Select original features using the aligned original indices
original_features_aligned = df.loc[original_indices_temporal, relevant_features]

# Concatenate temporal features with original features
X_meta = pd.concat([temporal_features_df, original_features_aligned], axis=1)
y_meta = df.loc[original_indices_temporal, 'stress_level'] # Use original stress levels


print("Meta input shape:", X_meta.shape)


# ===============================================
#  8. Train XGBoost Meta-Predictor
# ===============================================

from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Assign test sets to clearly named variables for access in later cells
X_meta_train_pipeline, X_meta_test_pipeline, y_meta_train_pipeline, y_meta_test_pipeline = train_test_split(X_meta, y_meta, test_size=0.2, random_state=42)

# Make these test sets globally accessible for prediction and evaluation in other cells
global X_meta_test_pipeline
global y_meta_test_pipeline

xgb_model = XGBRegressor(random_state=42)
xgb_model.fit(X_meta_train_pipeline, y_meta_train_pipeline)
preds = xgb_model.predict(X_meta_test_pipeline)

mse = mean_squared_error(y_meta_test_pipeline, preds)
rmse = np.sqrt(mse)
print(f"XGBoost Meta-Predictor  MSE: {mse:.4f}, RMSE: {rmse:.4f}")


# ===============================================
#  9. Summary
# ===============================================

print("\n--- Summary of Multi-Stage Pipeline ---")
print(f"Pipeline (Regression)  RMSE: {rmse:.4f}")
print("This pipeline uses an Autoencoder for feature compression, a Temporal GRU with Attention for sequence processing, and an XGBoost Regressor for final stress score prediction.")
print("The RMSE indicates the typical error in the predicted continuous stress score.")
```

## Comparison using a Common Classification Metric

To find a common comparison metric between the Decision Tree (classification) and XGBoost (regression) models, we can convert the continuous stress level predictions from the XGBoost pipeline into discrete integer stress levels by rounding. We can then compare both models using a classification metric like Accuracy.

**Note:** This approach allows for a direct comparison using a single metric, but it's important to remember that the XGBoost model was optimized for predicting a continuous value (minimizing MSE), not for maximizing classification accuracy on rounded values.

```python
from sklearn.metrics import accuracy_score
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Get the true discrete test labels for classification models
# Use y_test_clf_discrete which was explicitly created with integer labels
y_test_discrete_clf = y_test_clf_discrete


# Get the discrete predictions from the Decision Tree model
dt_discrete_pred = dt_pred

# Get the continuous predictions from the XGBoost pipeline (globally accessible)
# pipeline_predictions

# Convert the continuous XGBoost predictions to discrete integer stress levels by rounding
xgb_discrete_pred = np.round(pipeline_predictions).astype(int)

# Ensure the rounded XGBoost predictions are within the valid range of stress levels (1-10 based on original data insights)
# Although the original data was scaled, the predicted scaled values might fall outside the original integer range when rounded.
# Let's clip the rounded predictions to a reasonable range if necessary, based on the original stress level range (assuming 1-10).
# We need to check the original stress level range from the initial data inspection or summary statistics if available.
# From the describe(include='all') output in cell ec107fb2, the min stress level is 1.0 and max is 10.0.
# Let's check the range of predictions and clip if needed.
print("Original stress level range (from data description): 1 to 10")
if xgb_discrete_pred.size > 0:
    print(f"Min rounded XGBoost prediction: {xgb_discrete_pred.min()}")
    print(f"Max rounded XGBoost prediction: {xgb_discrete_pred.max()}")
else:
    print("XGBoost predictions array is empty.")


# Clip rounded predictions to the range of 1 to 10 if they fall outside this range
# This helps ensure the discrete predictions are comparable to the original stress levels.
xgb_discrete_pred_clipped = np.clip(xgb_discrete_pred, 1, 10)
if xgb_discrete_pred_clipped.size > 0:
    print(f"Min clipped XGBoost prediction: {xgb_discrete_pred_clipped.min()}")
    print(f"Max clipped XGBoost prediction: {xgb_discrete_pred_clipped.max()}")
else:
    print("Clipped XGBoost predictions array is empty.")


# Identify common indices between classification test set and pipeline test set
classification_test_indices = y_test_discrete_clf.index
pipeline_test_indices = y_meta_test_pipeline.index
common_indices = classification_test_indices.intersection(pipeline_test_indices)

print(f"\nNumber of samples in classification test set: {len(classification_test_indices)}")
print(f"Number of samples in pipeline test set: {len(pipeline_test_indices)}")
print(f"Number of common samples for comparison: {len(common_indices)}")

if len(common_indices) == 0:
    print("\nNo common samples found between classification and pipeline test sets. Cannot perform common metric comparison.")
else:
    # Filter true labels and predictions to include only common samples
    dt_true_common = y_test_discrete_clf.loc[common_indices].astype(int) # Ensure discrete and clip
    dt_true_common = np.clip(dt_true_common, 1, 10)

    # Ensure dt_discrete_pred is a pandas Series with the correct index for .loc
    dt_pred_series = pd.Series(dt_discrete_pred, index=y_test_clf_discrete.index)
    dt_pred_common = dt_pred_series.loc[common_indices]

    # True discrete labels for the pipeline test set (filtered to common indices)
    # y_meta_test_pipeline contains the true continuous values for the pipeline test set
    # Convert these to integers for comparison with rounded predictions
    xgb_true_common = y_meta_test_pipeline.loc[common_indices].astype(int) # Ensure discrete and clip
    xgb_true_common = np.clip(xgb_true_common, 1, 10)


    # Filter the rounded XGBoost predictions to common indices
    # xgb_discrete_pred_clipped is a numpy array, need to convert to Series for .loc
    xgb_pred_series = pd.Series(xgb_discrete_pred_clipped, index=y_meta_test_pipeline.index)
    xgb_pred_common = xgb_pred_series.loc[common_indices]

    # Diagnostic prints before accuracy calculation
    print("\nDiagnostic Info before Accuracy Calculation:")
    print(f"dt_true_common dtype: {dt_true_common.dtype}, unique values: {np.unique(dt_true_common)}")
    print(f"dt_pred_common dtype: {dt_pred_common.dtype}, unique values: {np.unique(dt_pred_common)}")
    print(f"xgb_true_common dtype: {xgb_true_common.dtype}, unique values: {np.unique(xgb_true_common)}")
    print(f"xgb_pred_common dtype: {xgb_pred_common.dtype}, unique values: {np.unique(xgb_pred_common)}")


    # Calculate Accuracy on common samples
    dt_accuracy_common = accuracy_score(dt_true_common.astype(int), dt_pred_common.astype(int))
    xgb_accuracy_common = accuracy_score(xgb_true_common.astype(int), xgb_pred_common.astype(int))


    print("\nCommon Metric Comparison (Accuracy on Discrete Stress Levels):")
    print(f"Decision Tree (Classification) - Accuracy (on common samples): {dt_accuracy_common:.4f}")
    print(f"XGBoost Pipeline (Regression - Rounded Predictions) - Accuracy (on common samples): {xgb_accuracy_common:.4f}")


    # Create a bar chart for Accuracy comparison
    comparison_data = {
        'Model': ['Decision Tree (Classification)', 'XGBoost Pipeline (Rounded Predictions)'],
        'Accuracy': [dt_accuracy_common, xgb_accuracy_common]
    }
    comparison_df = pd.DataFrame(comparison_data)

    # --- Enhance the bar chart ---
    plt.figure(figsize=(9, 6)) # Slightly larger figure size
    sns.barplot(x='Model', y='Accuracy', data=comparison_df.sort_values(by='Accuracy', ascending=False),
                palette='viridis', hue='Model', legend=False) # Use 'viridis' palette, remove legend
    plt.title('Accuracy Comparison on Discrete Stress Levels (on Common Samples)', fontsize=14) # Larger title font
    plt.ylabel('Accuracy', fontsize=12) # Larger ylabel font
    plt.xlabel('Model', fontsize=12) # Add xlabel
    plt.ylim(0, max(comparison_df['Accuracy']) * 1.1) # Adjust y-axis limit for better visual
    plt.xticks(rotation=0, ha='center', fontsize=10) # Center x-axis labels
    plt.grid(axis='y', linestyle='--', alpha=0.7) # Add horizontal grid lines

    # Add data labels on top of bars
    for index, row in comparison_df.sort_values(by='Accuracy', ascending=False).iterrows():
        plt.text(index, row['Accuracy'] + 0.01, f"{row['Accuracy']:.4f}", color='black', ha="center", fontsize=10)

    plt.tight_layout() # Adjust layout to prevent labels overlapping
    plt.show()
```

## Project Summary

This project involved analyzing a dataset titled "Digital Twin of Stress  Daily Journal.csv" with the goal of understanding and predicting stress levels based on various health and lifestyle factors.

The key steps undertaken were:

1.  **Data Loading and Initial Inspection:** The dataset was loaded and initially inspected to understand its structure and content.
2.  **Data Cleaning and Imputation:** Missing values were handled, and data was cleaned, including renaming and standardizing column names. Duplicates were dropped, and outliers were handled for numerical features.
3.  **Feature Engineering:** New features, `Sleep_Efficiency` and `Activity_to_Stress_Ratio`, were created to capture more complex relationships in the data.
4.  **Encoding and Scaling:** Categorical features were encoded using Label Encoding, and numerical features were scaled using StandardScaler to prepare the data for modeling.
5.  **Exploratory Data Analysis (EDA):** Univariate and bivariate analyses were performed using various visualizations (histograms, box plots, hexbin plots, scatter plots, correlation heatmaps) to understand the distributions of individual features, relationships between pairs of features, and correlations.
6.  **Model Building and Evaluation:**
    *   **Classification Models:** Logistic Regression, Decision Tree Classifier, and Random Forest Classifier were trained and evaluated using Accuracy and Macro F1-Score.
    *   **Multi-stage Pipeline:** A pipeline consisting of an Autoencoder for feature compression, a Temporal GRU with Attention for temporal feature extraction, and an XGBoost Regressor for final continuous stress score prediction was implemented and evaluated using regression metrics (MSE and RMSE).
    *   A comparison using a common classification metric (Accuracy after rounding XGBoost predictions) was also performed on aligned test samples.
7.  **Results Visualization and Inference:** Model performance was visualized, and inferences were drawn from the evaluation metrics and visualizations.
8.  **Export:** The cleaned dataset and model evaluation scores were saved to CSV files.

**Key Findings:**

*   Several features showed interesting distributions and relationships, although strong linear correlations between most original features were generally weak.
*   The newly engineered features provided potentially insightful combinations of existing data.
*   The classification models showed relatively high accuracy, with the Decision Tree performing best in terms of Macro F1-Score.
*   The multi-stage regression pipeline provided continuous stress score predictions with a certain level of error indicated by the RMSE.
*   Comparing classification and regression models requires careful consideration of their different output types and evaluation metrics, or using common metrics on transformed predictions with appropriate caveats.

The project successfully processed the data and built different types of models to predict stress levels, providing insights into the factors influencing stress based on the provided dataset.
